## 2.Which country has maximum no of AI Companies

```python
country_counts = df["Headquarters"].value_counts()
top_country = country_counts.idxmax()  # Gets the country with the highest count
max_count = country_counts.max()  # Gets the count of AI companies in that country
print(f"The country with the most AI companies is {top_country} with {max_count} companies.")
```

## 7.Compare the Annual Revenue and Glassdoor Score of older companies with new companies

```python
# Split the dataset into older and newer companies
older_companies = df[df["Founded"] < 2015]
newer_companies = df[df["Founded"] >= 2015]

# Get the statistics for older and newer companies
older_summary = older_companies[["Glassdoor Score", "Annual Revenue"]].describe()
newer_summary = newer_companies[["Glassdoor Score", "Annual Revenue"]].describe()

# Combine the statistics for comparison
comparison_table = pd.concat([older_summary, newer_summary], axis=1)

# Rename the columns to indicate 'Older Companies' and 'Newer Companies'
comparison_table.columns = ['Older Companies - Glassdoor Score','Older Companies - Annual Revenue','Newer Companies - Glassdoor Score', 'Newer Companies - Annual Revenue']

# Print the comparison table
comparison_table
```

## 8. What are the companies are having the best performance in past 5 years?

```python
# Filter companies founded in the past 5 years (2018â€“2024)
recent_companies = df[(df["Founded"] >= 2018) & (df["Founded"] <= 2024)]

# Sort by Glassdoor Score and Annual Revenue (you can prioritize one over the other)
best_performers = recent_companies.sort_values(by=["Glassdoor Score", "Annual Revenue"], ascending=False)

# Display the top 10 best-performing companies in the last 5 years
best_performers
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=recent_companies["Founded"], y=best_performers["Glassdoor Score"], palette="coolwarm",hue=best_performers["Annual Revenue"])
plt.xlabel("Year Founded")
plt.ylabel("Glassdoor Score")
plt.title("Year Founded vs Glassdoor Score")
plt.show()
```
