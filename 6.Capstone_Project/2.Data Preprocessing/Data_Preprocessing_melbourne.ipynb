```python
df=dataset.copy()
```

```python
df.drop(['Lattitude','Longtitude','Address','BuildingArea','Method','SellerG'],axis=1,inplace=True)
```

```python
df.shape
```

```python
df = df.dropna(subset=['YearBuilt'])
```

```python
df['YearBuilt'].value_counts()
```

```python
df.fillna('NA',inplace=True)
```

```python
df['CouncilArea'].value_counts()
```

```python
df['Car'].value_counts()
```

```python
#Handling the missing values in CouncilArea column
df['CouncilArea'].replace(to_replace=['NA'],value='Melbourne',inplace=True)
df['CouncilArea'].replace(to_replace=['Unavailable'],value='Melbourne',inplace=True)
```

```python
#Handling the Nan Values in Car column
df['Car'] = pd.to_numeric(df['Car'], errors='coerce').astype('Int64')
df['Car'] = df['Car'].fillna(df['Car'].mode()[0])
```

```python
df.isnull().sum()
```

```python
# Ensure Postcode is integer
df['Postcode'] = df['Postcode'].astype('int64')
df['Propertycount'] = df['Propertycount'].astype('Int64')
```

```python
df = df.drop_duplicates()
```
